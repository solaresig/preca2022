```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import glob
import os
from os import listdir
from os.path import isfile, join
import csv

dirlist=os.listdir('~/preca2022/original')
dirlist2=[]
dirlist3=[]
path0='~/preca2022/original/'
path2='/*.csv'
path3='.csv'
for i in range( 0, len(dirlist)):
    x=path0+dirlist[i]+path2
    dirlist2.append(x)

for i in range(0, len(dirlist2)):
    filelist=glob.glob(dirlist2[i])
    r1=re.compile(r"^(?!.*Tabla)")
    r2 = re.compile(".*abla_Apoyos")
    r3 = re.compile(".*abla_Bonos")
    r4 = re.compile(".*abla_Comisiones")
    r5 = re.compile(".*abla_Dietas")
    r6 = re.compile(".*abla_Estimulos")
    r7 = re.compile(".*abla_Gratific")
    r8 = re.compile(".*abla_Ingreso")
    r9 = re.compile(".*abla_Percep.*dinero")
    r10 = re.compile(".*Tabla_Prest.*econ")
    r11 = re.compile(".*Tabla_Primas")
    r12 = re.compile(".*Tabla_Siste")
    principal= list(filter(r1.match, filelist))
    apoyos=list(filter(r2.match, filelist)) 
    bonos=list(filter(r3.match, filelist)) 
    comis=list(filter(r4.match, filelist)) 
    dieta=list(filter(r5.match, filelist)) 
    estim=list(filter(r6.match, filelist)) 
    grati=list(filter(r7.match, filelist)) 
    ingres=list(filter(r8.match, filelist)) 
    perce=list(filter(r9.match, filelist)) 
    prest=list(filter(r10.match, filelist)) 
    prima=list(filter(r11.match, filelist)) 
    siste=list(filter(r12.match, filelist)) 
    with open(principal[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        row1=next(reader_obj)
        uni=row1[1]
        next(reader_obj)
        next(reader_obj)
        prin = pd.DataFrame(reader_obj)
        prin.columns = prin.iloc[0] 
        prin = prin[1:]
        prin = prin[['ID', 'EJERCICIO', 'DENOMINACIÓN DEL CARGO', 'NOMBRE (S)', 'PRIMER APELLIDO', 'SEGUNDO APELLIDO', 'SEXO (CATÁLOGO)', 'MONTO MENSUAL NETO DE LA REMUNERACIÓN  EN TABULADOR']]
        prin['Universidad']=uni
    with open(apoyos[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        ap = pd.DataFrame(reader_obj)
        ap = ap[1:]
        ap = ap.iloc[:, [0, 5, 7]]
        ap.columns = ['id', 'neto', 'periodicidad']
        ap['tipo']='apoyos'
    with open(bonos[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        bo = pd.DataFrame(reader_obj)
        bo = bo[1:]
        bo = bo.iloc[:, [0, 5, 7]]
        bo.columns = ['id', 'neto', 'periodicidad']
        bo['tipo']='bonos'
    with open(comis[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        co = pd.DataFrame(reader_obj)
        co = co[1:]
        co = co.iloc[:, [0, 5, 7]]
        co.columns = ['id', 'neto', 'periodicidad']
        co['tipo']='comisiones'
    with open(dieta[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        di = pd.DataFrame(reader_obj)
        di = di[1:]
        di = di.iloc[:, [0, 5, 7]]
        di.columns = ['id', 'neto', 'periodicidad']
        di['tipo']='dieta'
    with open(estim[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        es = pd.DataFrame(reader_obj)
        es = es[1:]
        es = es.iloc[:, [0, 5, 7]]
        es.columns = ['id', 'neto', 'periodicidad']
        es['tipo']='estimulos'
    with open(grati[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        gr = pd.DataFrame(reader_obj)
        gr = gr[1:]
        gr = gr.iloc[:, [0, 5, 7]]
        gr.columns = ['id', 'neto', 'periodicidad']
        gr['tipo']='gratificaciones'
    with open(ingres[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        ig = pd.DataFrame(reader_obj)
        ig = ig[1:]
        ig = ig.iloc[:, [0, 5, 7]]
        ig.columns = ['id', 'neto', 'periodicidad']
        ig['tipo']='ingresos'
    with open(perce[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        pe = pd.DataFrame(reader_obj)
        pe = pe[1:]
        pe = pe.iloc[:, [0, 5, 7]]
        pe.columns = ['id', 'neto', 'periodicidad']
        pe['tipo']='percepciones'
    with open(prest[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        pr = pd.DataFrame(reader_obj)
        pr = pr[1:]
        pr = pr.iloc[:, [0, 5, 7]]
        pr.columns = ['id', 'neto', 'periodicidad']
        pr['tipo']='prestaciones'
    with open(prima[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        pi = pd.DataFrame(reader_obj)
        pi = pi[1:]
        pi = pi.iloc[:, [0, 5, 7]]
        pi.columns = ['id', 'neto', 'periodicidad']
        pi['tipo']='primas'
    with open(siste[0]) as file_obj:
        reader_obj = csv.reader(file_obj)
        next(reader_obj)
        si = pd.DataFrame(reader_obj)
        si = si[1:]
        si = si.iloc[:, [0, 5, 7]]
        si.columns = ['id', 'neto', 'periodicidad']
        si['tipo']='sistemas'
    secundario=pd.concat([si, pi, pe, ig, gr, pr, es, di, co, bo, ap])
    secundario['Universidad']=uni
    path1='C:/Users/lito_/precarios/sumas/'
    sec='sec'
    pri='pri'
    a=path1+pri+dirlist[i]+path3
    b=path1+sec+dirlist[i]+path3
    prin.to_csv(a, index = False, encoding='latin1', errors='ignore')
    secundario.to_csv(b, index = False, encoding='latin1')


lista=glob.glob(r'~/preca2022/sumas/*.csv')
for i in range(0, len(lista)):
    r1=re.compile(".*pri.*")
    r2 = re.compile(".*sec.*")
    principal= list(filter(r1.match, lista))
    apoyos=list(filter(r2.match, lista))

general= pd.DataFrame()
extras = pd.DataFrame()
for i in range(0, len(principal)):
    reader = pd.read_csv(principal[i], encoding="latin1")
    general=pd.concat([general, reader])
for i in range(0, len(apoyos)):
    reader2 = pd.read_csv(apoyos[i], encoding="latin1")
    extras=pd.concat([extras, reader2])
general.to_csv('~/preca2022/sumas/general.csv', index = False, encoding='latin1', errors='ignore')
extras.to_csv('~/preca2022/sumas/extras.csv', index = False, encoding='latin1', errors='ignore')
```
